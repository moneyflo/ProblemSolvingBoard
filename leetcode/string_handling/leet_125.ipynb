## 125. Valid Palindrome
주어진 문자열이 팰린드롬인지 확인. 대소문자를 구분하지 않고, 영문자와 숫자만을 대상으로함.

### My Sol
Runtime: 64 ms, faster than 67.69% of Python3 online submissions for Valid Palindrome.

Memory Usage: 14.4 MB, less than 57.28% of Python3 online submissions for Valid Palindrome.

```python
def isPalindrome(self, s: str) -> bool:
    alp = 'abcdefghijklmnopqrstuvwxyz0123456789'
        
    s = s.replace(' ', '') # 밑에서 확인해주니까 필요없을듯
    s = s.lower()
        
    sol = ''
        
    for x in s:
        if x in alp:
            sol += x
        
    if sol == sol[::-1]:
        return True
    else:
        return False
```

### 풀이 1 : 리스트로 변환
Runtime: 665 ms, faster than 5.03% of Python3 online submissions for Valid Palindrome.

Memory Usage: 19.2 MB, less than 11.80% of Python3 online submissions for Valid Palindrome.

```python
def isPalindrome(s: str) -> bool:
    # 대소문자 여부를 무시하고 영문자, 숫자만을 대상으로 한다는 조건에 대한 전처리
    strs = []
    for char in s:
        if char.isalnum():   # 영문자, 숫자 여부를 판별하는 함수
            strs.append(char.lower())

    # 팰린드롬 여부 판별
    while len(strs) > 1:
        if strs.pop(0) != strs.pop():
            return False
    
    return True
```

### 풀이 2 : 데크 자료형을 이용한 최적화
Runtime: 72 ms, faster than 54.78% of Python3 online submissions for Valid Palindrome.

Memory Usage: 19.3 MB, less than 6.28% of Python3 online submissions for Valid Palindrome.

```python
import collections

def isPalindrome(s: str) -> bool:
    # 자료형 데크로 선언
    strs: Deque = collections.deque()

    for char in s:
        if char.isalnum():
            strs.append(char.lower())
    
    while len(strs) > 1:
        if strs.popleft() != strs.pop():
            return False

    return True
```

리스트의 pop(0)는 O(n)이지만, 데크의 popleft()는 O(1)이기 때문에<br/>
각각 n번씩 반복하면 리스트 구현은 O(n^2), 데크 구현은 O(n)

### 풀이 3 : 슬라이싱 사용
Runtime: 76 ms, faster than 48.39% of Python3 online submissions for Valid Palindrome.

Memory Usage: 15.6 MB, less than 17.90% of Python3 online submissions for Valid Palindrome.

```python
import re

def isPalindrome(s: str) -> bool:
    s = s.lower()
    # 정규식으로 불필요한 문자 필터링
    s = re.sub('[^a-z0-9]', '', s)

    return s == s[::-1] # 슬라이싱
```

```python
isPalindrome("A man, a plan, a canal: Panama")
```

문자열 슬라이싱은 내부적으로 매우 빠르게 동작함.<br/>
따라서 문자열을 조작할 때는 항상 슬라이싱을 우선으로 사용하는 편이 속도 개선에 유리함.<br/>
별도의 자료형으로 매핑하면 그 과정에서 상당한 연산 비용이 필요하므로 속도가 느려짐.

* 슬라이싱 0.499 ms ( = 1 slice)
* 리스트 reverse()      5
* for 반복              12
* while                 21
* 재귀                  54
